```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
file = r'1.01. Simple linear regression.csv'
data = pd.read_csv(file)
x = data.set_index(data['SAT'])
y = data.set_index(data['GPA'])
```

```python
plt.scatter(x, y)
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
data.hist()
plt.show()
```

```python
regression_model = linear_model.LinearRegression()
regression_model.fit(x, y)
print(regression_model.intercept_)
print(regression_model.coef_)
```

```python
file = r'mtcars.csv'
data = pd.read_csv(file)
```

```python
correlation = data.corr()
data.hist()
print(correlation)
plt.show()
```

```python
X = data.drop(data.columns[[0, 2]], axis= 1)
data.plot(kind= 'scatter', x = 'wt', y = 'mpg', figsize= (9, 9), color= 'black')
plt.show()
```

```python
from sklearn import linear_model

X = pd.DataFrame(data['wt'])
y = data['mpg']

model = linear_model.LinearRegression()
model.fit(X,y)
print(model.intercept_)
print(model.coef_)
model.score(X,y)


y_pred = model.predict(X)
residual = data['mpg']-y_pred
residual.describe()
```

```python
data.plot(kind='scatter',x = 'wt',y = 'mpg',figsize=(9,9),color='black')
plt.plot(data['wt'],y_pred,color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt'] ** 2]).T
y = data['mpg']
model.fit(X, y)
print('Intercept cua mo hinh bac 2: ')
print(model.intercept_)
print('Coefficient cua mo hinh bac 2: ')
print(model.coef_)
print('Do chinh xac cua mo hinh bac 2:')
print(model.score(X, y))
```

```python
import numpy as np
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range, range ** 2]).T
y_pred = model.predict(X = predictor)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize= (9, 9), color = 'black', xlim= (0, 7))
plt.plot(range, y_pred, color = 'blue')
```
