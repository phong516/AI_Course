```python
#write to file
with open('text.txt', mode = 'w') as accounts:
  accounts.write('201461  Nguyễn A 10\n')
  accounts.write('201462  Nguyễn B  6\n')
  accounts.write('201463  Nguyễn C  4\n')
  print('201464  Nguyễn D  2', file = accounts)
```

```python
#read file
with open('text.txt', mode = 'r') as accounts:
  print(f"{'MSSV':< 10}{'Name' :< 20}{'Score' :> 2}")
  for record in accounts:
    mssv, name, score = record.split()
    print(f'{mssv :< 10}{name :< 20}{score :> 2}')
```

```python
import pandas as pd
url = 'http://winterolympicsmedals.com/medals.csv'
data = pd.read_csv(url)

```

```python
data_medal = data[(data['NOC'] == 'CAN')]
print("Number of Medals by CAN: "+ str(len(data_medal)))
```

```python
data_gold = data[(data['Event'] == 'individual') & (data['Medal'] == 'Gold')]
print("Number of Gold Medals by Individual Event: "+ str(len(data_gold)))
```
